#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
# Склеиваем таблицы информации по строкам.
df1_1 = pd.read_csv("Data/Students_info_0.csv", sep=",")
for i in range(1,10):
  df1_2 = pd.read_csv("Data/Students_info_" + str(i) + ".csv", sep=",")
  df1_1 = pd.concat([df1_1, df1_2])

# Склеиваем таблицы оценок по строкам.
df2_1 = pd.read_csv("Data/Students_marks_0.csv", sep=",")
for i in range(1,10):
  df2_2 = pd.read_csv("Data/Students_marks_" + str(i) + ".csv", sep=",")
  df2_1 = pd.concat([df2_1, df2_2])

# Объединяем таблицу информации и таблицу оценок по индексу.
df = pd.merge(df1_1, df2_1, how='left', on='index')
df.sample(n=10)
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
# Создаем столбцы результатов по каждому предмету (используем лямбда-функцию для проверки результата за экзамен).
df["Math_PassStatus"] = df["math score"].apply(lambda res: "F" if res < 50 else "P")
df["Reading_PassStatus"] = df["reading score"].apply(lambda res: "F" if res < 50 else "P")
df["Writing_PassStatus"] = df["writing score"].apply(lambda res: "F" if res < 50 else "P")

# Считаем количество студентов, сдавших и не сдавших экзамен по каждому предмету.
print(df["Math_PassStatus"].value_counts().to_frame('Math_PassStatus'))
print(df["Reading_PassStatus"].value_counts().to_frame('Reading_PassStatus'))
print(df["Writing_PassStatus"].value_counts().to_frame('Writing_PassStatus'))
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
# Функция для проверки условия.
def GetResult(x):
  if x['Math_PassStatus'] == "F" or x['Reading_PassStatus'] == "F" or x['Writing_PassStatus'] == "F": return "F"
  else: return "P"

# Создаем столбец итоговых результатов.
df["OverAll_PassStatus"] = df.apply(GetResult, axis=1)

# Считаем количество студентов, сдавших все экзамены.
print(df["OverAll_PassStatus"].value_counts()["P"])
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
  if average_mark > 90: return "A"
  elif average_mark > 80: return "B"
  elif average_mark > 70: return "C"
  elif average_mark > 60: return "D"
  elif average_mark > 50: return "E"
  else: return "F"

# Создаем столбец итоговых результатов.
df["Grade"] = df[['math score','reading score','writing score']].mean(axis = 1).apply(GetGrade)

# Выводим количество студентов, получивших каждую из оценок.
df["Grade"].value_counts()
```
